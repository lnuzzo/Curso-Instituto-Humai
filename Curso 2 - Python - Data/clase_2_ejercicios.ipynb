```python
# ¿Cuál es la producción total en toneladas estimada de todos los cultivos en la Argentina, cada año?
produccion_anual = estimaciones[estimaciones['campania_fin'] < 2019].pivot_table(
    index="campania_fin",
    values="produccion",
    aggfunc="sum"
)
produccion_anual
```

```python
produccion_anual.plot(figsize=(15,10))
```

```python
registros_por_anio = estimaciones.pivot_table(
    # index="anio",
    index = ['campania_fin'],
    values="produccion",
    aggfunc="count"
)
registros_por_anio
```

```python
registros_por_anio.plot(figsize=(15,10))
```

```python
distribucion_soja = estimaciones[
                                 (estimaciones.cultivo=='Soja total') & 
                                 (estimaciones.rendimiento<20000) &
                                 (estimaciones.id_departamento!='000')
                                ].pivot_table(
    index=['campania_fin','id_departamento','departamento'],
    values=['sup_sembrada','sup_cosechada','produccion','rendimiento'],
    aggfunc='sum'
).reset_index()
distribucion_soja['rendimiento'].max()
```

```python
# ¿Cómo se distribuye el rendimiento (kg por hectárea) estimado promedio del cultivo de soja entre los departamentos que la siembran? 
distribucion_soja_sfe = estimaciones[(estimaciones.cultivo=='Soja total') &
    (estimaciones.provincia=='SANTA FE')].pivot_table(
    index='campania_fin',
    columns=['id_departamento','departamento'],
    values='rendimiento',
    aggfunc='sum'
)
distribucion_soja_sfe.plot(figsize=(15,10))
```

```python
# box and whisker plots
distribucion_soja.plot(kind='box', subplots=True, figsize=[30,10], sharex=False, sharey=False)
plt.show()
```

```python
# scatter plot matrix
scatter_matrix(distribucion_soja,figsize=[30,10])
plt.show()
```

```python
# ¿Qué tipo de distribución probabilística tiene? 
bin_values = np.arange(start=0, stop=6000, step=100)
distribucion_soja['rendimiento'].hist(bins=bin_values, figsize=[20,10])
# RTA: Tiene una distribución normal con un leve sesgo positivo (alargamiento de la cola de la curva hacia la derecha)
# Fenómeno "Cero inflado" muchoas casos de ceros informados
```
